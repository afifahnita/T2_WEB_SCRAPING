```python
# Download stopwords Indonesia dari GitHub langsung
!wget https://raw.githubusercontent.com/stopwords-iso/stopwords-id/master/stopwords-id.txt -O stopwords.txt
```

```python
import pandas as pd
import re

# Load stopwords
with open("stopwords.txt", "r", encoding="utf-8") as f:
    stopwords = set(word.strip() for word in f.readlines())

# Ganti 'namafile.csv' dan kolom sesuai kebutuhanmu
df = pd.read_csv("/content/tweets-data/acab.csv")  # ← ganti sesuai nama file kamu
kolom_teks = "favorite_count"         # ← ganti dengan nama kolom yang mau dibersihkan

# Fungsi cleaning
def clean_text(teks):
    teks = str(teks).lower()
    teks = re.sub(r"http\S+|www\S+", "", teks)        # Hapus URL
    teks = re.sub(r"@\w+|#\w+", "", teks)              # Hapus mention & hashtag
    teks = re.sub(r"[^a-z\s]", "", teks)               # Hapus angka & simbol
    kata = teks.split()
    kata = [k for k in kata if k not in stopwords]     # Hapus stopwords
    return " ".join(kata)

# Terapkan cleaning
df["hasil_clean"] = df[kolom_teks].apply(clean_text)

# Simpan hasil ke file TXT (1 kalimat per baris untuk Wordij)
with open("clean_for_wordij.txt", "w", encoding="utf-8") as f:
    for line in df["hasil_clean"]:
        f.write(line + "\n")

print("✅ Data berhasil dibersihkan dan disimpan sebagai 'clean_for_wordij.txt'")
```
